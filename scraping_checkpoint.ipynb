```python
'''
Wrap all the previous functions into a single function that takes as parameters a Wikipedia link
'''
import requests
from bs4 import BeautifulSoup

def get_and_parse_html(url):
    # Get the HTML content from the URL
    response = requests.get(url)
    # Parse the HTML content using Beautiful Soup
    soup = BeautifulSoup(response.content, 'html.parser')
    return soup

def extract_article_title(soup):
    # Extract the article title using the h1 tag
    title = soup.find('h1').text
    return title

def extract_article_text(soup):
    # Initialize an empty dictionary to store the article text
    article_text = {}
    # Find all the headings (h2, h3, h4, etc.) and paragraphs
    headings = soup.find_all(['h2', 'h3', 'h4', 'h5', 'h6'])
    paragraphs = soup.find_all('p')
    # Iterate through the headings and paragraphs
    for heading in headings:
        # Extract the heading text and store it in the dictionary as the key
        heading_text = heading.text
        article_text[heading_text] = []
        # Find the next sibling of the heading that is a paragraph
        next_sibling = heading.find_next_sibling('p')
        while next_sibling:
            # Extract the paragraph text and store it in the dictionary as the value
            paragraph_text = next_sibling.text
            article_text[heading_text].append(paragraph_text)
            # Find the next sibling of the paragraph that is a paragraph
            next_sibling = next_sibling.find_next_sibling('p')
    # Iterate through the paragraphs
    for paragraph in paragraphs:
        # Extract the paragraph text and store it in the dictionary as the key
        paragraph_text = paragraph.text
        article_text[paragraph_text] = []
    return article_text

def extract_links(soup):
    # Initialize an empty list to store the links
    links = []
    # Find all the anchor tags
    anchor_tags = soup.find_all('a')
    # Iterate through the anchor tags
    for tag in anchor_tags:
        # Extract the link and check if it redirects to another Wikipedia page
        link = tag.get('href')
        if link.startswith('/wiki/'):
            links.append(link)
    return links

def scrape_wikipedia(url):
    soup = get_and_parse_html(url)
    title = extract_article_title(soup)
    article_text = extract_article_text(soup)
    links = extract_links(soup)
    return title, article_text, links

# Test the function on a Wikipedia page of your choice
url = 'https://en.wikipedia.org/wiki/Beautiful_Soup_(HTML_parser)'
title, article_text, links = scrape_wikipedia(url)
print(title)
print(article_text)
print(links)

```
